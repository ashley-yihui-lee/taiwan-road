## Purpose of this notebook: merge all cities and counties together and do data analysis on Taiwan's road names.

```python
import pandas as pd

df_taipei = pd.read_csv("臺北市_roads.csv")
df_taipei["city"] = "臺北市"

df_newtaipei = pd.read_csv("新北市_roads.csv")
df_newtaipei["city"] = "新北市"

df_taoyuan = pd.read_csv("桃園市_roads.csv")
df_taoyuan["city"] = "桃園市"

df_hsinchu_city = pd.read_csv("新竹市_roads.csv")
df_hsinchu_city["city"] = "新竹市"

df_hsinchu_county = pd.read_csv("新竹縣_roads.csv")
df_hsinchu_county["city"] = "新竹縣"

df_miaoli = pd.read_csv("苗栗縣_roads.csv")
df_miaoli["city"] = "苗栗縣"

df_taichung = pd.read_csv("臺中市_roads.csv")
df_taichung["city"] = "臺中市"

df_changhua = pd.read_csv("彰化縣_roads.csv")
df_changhua["city"] = "彰化縣"

df_nantou = pd.read_csv("南投縣_roads.csv")
df_nantou["city"] = "南投縣"

df_yunlin = pd.read_csv("雲林縣_roads.csv")
df_yunlin["city"] = "雲林縣"

df_chiayi_city = pd.read_csv("嘉義市_roads.csv")
df_chiayi_city["city"] = "嘉義市"

df_chiayi_county = pd.read_csv("嘉義縣_roads.csv")
df_chiayi_county["city"] = "嘉義縣"

df_tainan = pd.read_csv("臺南市_roads.csv")
df_tainan["city"] = "臺南市"

df_kaohsiung = pd.read_csv("高雄市_roads.csv")
df_kaohsiung["city"] = "高雄市"

df_pingtung = pd.read_csv("屏東縣_roads.csv")
df_pingtung["city"] = "屏東縣"

df_yilan = pd.read_csv("宜蘭縣_roads.csv")
df_yilan["city"] = "宜蘭縣"

df_hualien = pd.read_csv("花蓮縣_roads.csv")
df_hualien["city"] = "花蓮縣"

df_taitung = pd.read_csv("臺東縣_roads.csv")
df_taitung["city"] = "臺東縣"

df_keelung = pd.read_csv("基隆市_roads.csv")
df_keelung["city"] = "基隆市"

df_penghu = pd.read_csv("澎湖縣_roads.csv")
df_penghu["city"] = "澎湖縣"

df_kinmen = pd.read_csv("金門縣_roads.csv")
df_kinmen["city"] = "金門縣"

df_lienchiang = pd.read_csv("連江縣_roads.csv")
df_lienchiang["city"] = "連江縣"

all_roads = pd.concat([
    df_taipei, df_newtaipei, df_taoyuan, df_hsinchu_city, df_hsinchu_county, df_miaoli,
    df_taichung, df_changhua, df_nantou, df_yunlin, df_chiayi_city, df_chiayi_county,
    df_tainan, df_kaohsiung, df_pingtung, df_yilan, df_hualien, df_taitung,
    df_keelung, df_penghu, df_kinmen, df_lienchiang
], ignore_index=True)
```

## Translation

```python
eng_name = pd.read_excel("translation.xls")
```

```python
eng_name = eng_name.rename(columns={
    "中文街路名稱": "mandarin",
    "英文街路名稱": "english"
})
eng_name
```

```python
merged = all_roads.merge(
    eng_name, 
    left_on="road", 
    right_on="mandarin", 
    how="left"
)

merged 
```

```python
merged.drop(columns=["mandarin"])
```

```python
from collections import Counter

#ignore the common words in road names (e.g. Lane, Section, Street...etc)
stop_words = ["Rd.", "St.", "Sec.", "Ln.", "Ln.,", "1st", "2nd", "3rd", "4th", "5th", "6th", "7th", "8th", "9th","10th",
              "1,", "2,", "3,", "4,", "E.", "S.", "N.", "W.", "Vil."]

all_words = " ".join(merged["english"].dropna()).split()

filtered_words = [word for word in all_words if word not in stop_words]

word_counts = Counter(filtered_words)

word_counts.most_common(20)
```
